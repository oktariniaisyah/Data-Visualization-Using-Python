# Data Visualization
In this project, I would like to show how I made a data visualization from table to graph. The aim of this project is to compare between the population growth of Zimbabwe and United States. So, I started to load data from csv file and analyzed the data.

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
#example of loading the data from sample_data.csv file
data = pd.read_csv('sample_data.csv')
```

```python
data
```

```python
plt.plot(data.column_a, data.column_b)
plt.plot(data.column_a, data.column_c)
plt.xlabel('X')
plt.ylabel('Y and Z')
plt.legend(['this is Y', 'this is Z'])
plt.show()
```

```python
#started to load the data from countries.csv file
table = pd.read_csv('countries.csv')
```

```python
table
```

```python
#compare population between Zimbabwe and US
```

```python
zimbabwe = table[table.country == 'Zimbabwe']
zimbabwe
```

```python
us = table[table.country == 'United States']
```

```python
us
```

```python
plt.plot(zimbabwe.year, zimbabwe.population / 10**6)
plt.plot(us.year, us.population / 10**6)
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['Zimbabwe','United States'])
plt.show()
```

From picture above, we can see that Zimbabwe's population is lower than United States's. At the first 1952, United States'population growing faster up to 150 Million until 301 M in 2007 than Zimbabwe started in 3 M until 12 M. Next, we're gonna see how the population growth in percentage between Zimbabwe and United States.

```python
zimbabwe.population
```

```python
zimbabwe.population / zimbabwe.population.iloc[0] * 100
```

```python
plt.plot(zimbabwe.year, zimbabwe.population / zimbabwe.population.iloc[0] * 100)
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.legend(['Zimbabwe','United States'])
plt.show()
```

From the graphic above, the population growth of Zimbabwe is higher than United States at the first 1950 until 2007. It's shows that Zimbabwe's population growth rapidly than United States even US has bigger population than Zimbabwe.
